Guardar y abrir un modelo <br>
-----------
<b>
Entrenar un modelo de Machine Learning es un proceso muchas veces lento, por lo que no tiene sentido volver a entrenar el modelo cada vez que lo necesitemos en el futuro.
Por suerte, una herramienta de SciKit Learn nos permite guardar nuestro modelo ya entrenado para utilizarlo cuando lo necesitemos.
</b>

<b>
1.- Guardar modelo:
</b>
Utilizaremos un directorio de SciKit Learn llamado “externals”, que contiene dependencias externas agrupadas que se actualizan de vez en cuando. En concreto utilizaremos joblib, una librería usada para realizar funciones de pipelining en Python. Importamos la librería de la siguiente forma: 

<code>
from sklearn.externals import joblib 
</code>

Ahora supongamos que tenemos el siguiente modelo, clf_rf, ya entrenado:

<code>
clf_rf.fit(x_train, y_train) # Entrenamiento del modelo 
</code>

Podemos guardarlo realizando un joblib.dump():

<code>
joblib.dump(clf_rf, 'modelo_entrenado.pkl') # Guardo el modelo.
</code>

El modelo se guardará en el fichero “modelo_entrenado.pkl” dentro del directorio que hayamos establecido por defecto en nuestro intérprete de Python.

<b>
2.- Cargar modelo:
</b>
Cuando necesitemos cargar el modelo ya entrenado, simplemente hacemos un joblib.load(): 

<code>
clf_rf = joblib.load('modelo_entrenado.pkl') # Carga del modelo.
</code>

Si queremos asegurarnos que el modelo se ha guardado correctamente, podemos calcular el rendimiento del modelo antes de guardarlo y al cargarlo de nuevo de la siguiente forma: <br>
<code>
clf_rf.score(x_train, y_train)
</code>

Ambos rendimientos deben ser exactamente iguales.

KNN : Nearest Neighbors Classification
-----------
<i>“Dime con quién vas y te diré quién eres”</i> <br>

El K-NN es un algoritmo de aprendizaje supervisado, es decir, que a partir de un juego de datos inicial su objetivo será el de clasificar correctamente todas las instancias nuevas. El juego de datos típico de este tipo de algoritmos está formado por varios atributos descriptivos y un solo atributo objetivo (también llamado clase).

El algoritmo clasifica cada dato nuevo en el grupo que corresponda, según tenga k vecinos más cerca de un grupo o de otro. Es decir, calcula la distancia del elemento nuevo a cada uno de los existentes, y ordena dichas distancias de menor a mayor para ir seleccionando el grupo al que pertenecer. Este grupo será, por tanto, el de mayor frecuencia con menores distancias.

Destacar que K-NN es muy sensible a:

- La variable k, de modo que con valores distintos de k podemos obtener resultados también muy distintos. Este valor suele fijarse tras un proceso de pruebas con varias instancias.
- La métrica de similitud utilizada, puesto que esta influirá, fuertemente, en las relaciones de cercanía que se irán estableciendo en el proceso de construcción del algoritmo. La métrica de distancia puede llegar a contener pesos que nos ayudarán a calibrar el algoritmo de clasificación, convirtiéndola, de hecho, en una métrica personalizada.

Su mayor debilidad es la lentitud en el proceso de clasificación puesto que su objetivo no es obtener un modelo optimizado, sino que cada instancia de prueba es comparada contra todo el juego de datos de entrenamiento y, será la bondad de los resultados lo que determinará el ajuste de aspectos del algoritmo como el propio valor k, el criterio de selección de instancias para formar parte del juego de datos “D” de entrenamiento o la propia métrica de medida de similitud.

Ejemplo
----
[Datos usados en este ejemplo](https://archive.ics.uci.edu/ml/datasets/car+evaluation) o puedes pinchar [aquí](https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data) para descargar directamente 

```python
import numpy as np
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('car.data')
```

```python
data.shape
```

```python
data.head()
```

```python
data = pd.read_csv('car.data',
                   header = None,
                   names = ['buying','maint','doors','persons','lug_boot','safety','class'],
                   index_col = False)
```

```python
data.shape
```

```python
data.head()
```

```python
x = data[['buying','maint','safety']].values
y = data[['class']]
```

```python
print(x,y)
```

<b> 
Hasta ahora se cargó la información de 'car.data' y se asignó el nombre de sus variables. <br>
Se usarán los atributos 'buying','main' y 'safety' como parámetros para el modelo KNN. Sin embargo, los atributos escogidos corresponden a variables categoricas, por lo que serán transformados usando <i>Label encoder</i>
</b>

```python
Le = LabelEncoder()
for i in range(len(x[0])): #len(x[0]) es una forma de referirse al número de columnas
    x[:,i] = Le.fit_transform(x[:, i])
```

```python
x
```

```python
y['class']
```

```python
label_mapping = {
    'unacc':0,
    'acc':1,
    'good':2,
    'vgood':3
}
y = y['class'].map(label_mapping)
y = np.array(y)
y
```

#### Uso de modelo

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=25,weights='uniform') # Define parámetros

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2) # Toma datos de testeo

knn.fit(x_train,y_train) # Se entrena modelo

prediction = knn.predict(x_test) # Resultado de predición

accuracy = metrics.accuracy_score(y_test,prediction) # Medida de desempeño (Max = 1)

print('Prediction: ',prediction)
print('Accuracy: ',accuracy)
```

<b>
    Con las métricas utilizadas se obtuvo un modelo que predice correctamente un 76% de los datos entregados (train + test).
    Los valores predichos se pueden consultar como: <code>knn.predict(x)[i]</code> 
    ,donde i corresponde al índice. Así se puede comparar con el dato real <code>y[i]</code>
    <br>
    A continuación se tomarán al azar 100 valores entre 0 y 1729, correspondiente a la cantidad de registros. El porcentaje de acertividad debiese rondar en con el porcentaje de Accuracy calculado previamente
</b>

```python
import random

correctas = 0
incorrectas = 0
n_muestras = 100
random_list = random.sample(range(0, 1729), n_muestras)
for i in random_list:
    if y[i] == knn.predict(x)[i]:
        correctas += 1
    else:
        incorrectas += 1
        
print('Buenas: ',correctas)
print('Malas: ',incorrectas)
print('Porcentaje correcto: ',(correctas/n_muestras)*100, '%' )
```

```python
num = 1003
print('Actual value: ', y[num])
print('Predict value: ',knn.predict(x)[num])
```
